**CHAPETR 17.2 LINEAR INTEPOLATION**

```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
```

```python
x = [1, 3, 5, 7]
y = [2, 8, 6, 4]

f = interp1d(x, y)
y_hat = f(3)
print(y_hat)
```

```python
plt.figure(figsize= (12,10))
plt.plot(x,y, '-ob')
plt.plot(3, y_hat, 'ro')
plt.title('Linear Interpolation at x = 3')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

**CHAPTER 17.3 CUBIC SPLINE INTEROLATION**

```python
from scipy.interpolate import CubicSpline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [0, 1, 2]
y = [1, 5, 2]

f = CubicSpline(x, y, bc_type='natural')
x_new = np.linspace(0, 2, 100)
y_new = f(x_new)
```

```python
plt.figure(figsize = (16,12))
plt.plot(x_new, y_new, 'b')
plt.plot(x, y, 'ro')
plt.title('Cubic Spline Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
